```python
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split ,StratifiedKFold , cross_val_score
from sklearn.metrics import classification_report, accuracy_score,make_scorer,f1_score
from imblearn.pipeline import Pipeline

#selecting feature and target
X = df.drop('Class',axis =1) #feature
y = df['Class'] # our target

# at here we split the data set first than make balance the dataset with smote

#now model training by 70 - 30 ratio
X_train, X_test , y_train ,y_test = train_test_split(X, y, test_size = 0.3,random_state = 48)

mypipeline = Pipeline([
    ('smote',SMOTE(random_state = 48)),
    ('rf',RandomForestClassifier(n_estimators = 70 ,max_depth = 8, random_state =48))
])
#use straight field fold for evenly balanced imbalanced data
skf = StratifiedKFold(n_splits=4 , shuffle=True ,random_state= 48)
# by cross value score model
scores= cross_val_score(mypipeline, X ,y ,cv =skf ,scoring = 'f1')
print("F1 score for each fold :", scores)
print('\n')
print("Average F1 score :",scores.mean())


smote = SMOTE(random_state = 48)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train ,y_train)

myModel = RandomForestClassifier(n_estimators=70 ,max_depth=8,random_state=48)
myModel.fit(X_train_resampled, y_train_resampled)


y_predic = myModel.predict(X_test)
print(classification_report(y_test , y_predic))
print('\n')
print('Accuracy score for model : ',accuracy_score(y_test,y_predic))
```
