```python
import numpy as np
import pandas as pd
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt
uploaded = files.upload()
df  = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isnull().sum()
df['Age'].fillna(df['Age'].median(),inplace = True)

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace = True)

df['Age'].isnull().sum()# checking there is no null value left
# now remove duplicates
df.duplicated().sum()
df.drop_duplicates(inplace = True)
```

```python
# now visualize age througt boxplot by seaborn
x= df['Age']
sns.boxplot(x)
plt.title('outlier in AGE col')
plt.show()
```

```python
# now categorize col by gender distribution
x = df['Sex'].value_counts()
sns.barplot(x)
plt.title('gender distribution')
plt.show()
df['Sex'].value_counts()
```

```python
histogram = df['Age']
histogram.hist()
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
# Correlation heatmap for numeric variables
# df.corr()
corr_numeric = df.select_dtypes(include = 'number')
sns.heatmap(corr_numeric.corr(),annot=True, cmap='coolwarm')
plt.title("Correlation Between Features")
plt.show()
```
